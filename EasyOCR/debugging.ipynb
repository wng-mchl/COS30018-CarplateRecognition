```python
import cv2
import torch
import os
from ultralytics import YOLO
```

```python
# Path to your trained model
MODEL_PATH = "/content/gdrive/My Drive/cos30018-test/yolov11/runs/detect/train5/weights/best.pt"

# Load the trained YOLOv8 model
model = YOLO(MODEL_PATH)

# Path to test image
image_path = "/content/gdrive/My Drive/cos30018-test/data/images/train/20250315022939_038836_frame_0000.jpg"

# Run inference
results = model(image_path)  # can directly pass the image path

# Annotate and save the image with bounding boxes
annotated_frame = results[0].plot()  # annotated image as NumPy array

# Save the result to disk
debug_output_path = "/content/gdrive/My Drive/cos30018-test/EasyOCR/PredictedImages/debug_output.jpg"
cv2.imwrite(debug_output_path, annotated_frame)

print(f"✅ Annotated image saved to: {debug_output_path}")
```
