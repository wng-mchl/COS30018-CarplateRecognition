```python
import os
from ultralytics import YOLO

model = YOLO(os.path.join(ROOT_DIR, "yolov11/train_110525/runs/detect/train2/weights/best.pt"))

results = model.train(
    data=os.path.join(ROOT_DIR, "yolov11/google_colab_config.yaml"),
    epochs=10,
    imgsz=640,           # image size
    batch=16,            # batch size
    lr0=0.0001,           # initial learning rate
    optimizer="AdamW",   # optimizer: SGD, Adam, AdamW
    weight_decay=0.0001,  # regularization
    warmup_epochs=3,     # warmup phase
    patience=10,         # early stopping
    workers=4,           # dataloader workers
    device=0             # or "cuda:0" for GPU
)
```
